## **Descripción de la situación matemática**

El problema que se aborda es la evaluación de funciones de optimización aplicadas al contexto de redes neuronales. En el entrenamiento de modelos de aprendizaje automático, especialmente redes neuronales, es común enfrentarse a funciones de costo o pérdida que deben minimizarse. Para estudiar y comparar algoritmos de optimización, se utilizan funciones matemáticas de prueba que presentan distintos grados de dificultad y características específicas.

**Función de Rosenbrock**

Es una función clásica en optimización que genera un valle estrecho y curvado.

**Función Sphere modificada**

Es una variante de la función cuadrática simple, donde se suma el cuadrado de cada variable.

```python
import math

# Función de Rosenbrock: prueba clásica en optimización
def rosen(xx):
    d = len(xx)
    suma = 0
    for i in range(d - 1):
        xi = xx[i]
        xnext = xx[i + 1]
        suma += 100 * (xnext - xi**2)**2 + (xi - 1)**2
    return suma

# Función Sphere modificada: incluye penalización si la suma excede 1
def spherefmod(xx):
    suma = sum(x**2 for x in xx)
    if sum(xx) > 1.0:
        suma += 10.0 * (sum(xx) - 1.0)**2
    return suma

# Función de Rastrigin: otra prueba clásica en optimización
def rastrigin(xx):
    d = len(xx)
    suma = 10 * d
    for x in xx:
        suma += x**2 - 10 * math.cos(2 * math.pi * x)
    return suma

# Bloque interactivo principal
if __name__ == "__main__":
    print("Simulación de funciones de optimización en REDES NEURONALES")
    valores = [float(x) for x in input("Ingresa valores separados por espacio: ").split()]

    print("Rosenbrock:", rosen(valores))
    print("Rastrigin:", rastrigin(valores))

    if len(valores) == 6:
        print("Sphere modificada:", spherefmod(valores))
```
