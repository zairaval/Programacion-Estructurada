```python
import math  # Biblioteca de funciones matemáticas

#Convertir de grados Celsius a Fahrenheit.
#Fórmula: F = (C * 9/5) + 32

def celsius_a_fahrenheit(celsius):
  #Parámetros: celsius (float)-> Temperatura en grados Celsius.
  #Retorna:float: Temperatura en grados ->Fahrenheit.
    return (celsius * 9/5) + 32


# Calcular el área de un círculo.
# Fórmula: A = π * r^2

def calcular_area_circulo(radio):

    #Parámetros: radio (float)-> Radio del círculo.
    #Retorna:float-> Área del círculo.

    if radio < 0:
        raise ValueError("El radio no puede ser negativo.")
    return math.pi * (radio ** 2)


#Determinar si un número es par.
def es_par(numero):

    #Determinar si un número es par.
    #Parámetros: numero (int)-> Número entero.
    #Retorna:bool-> True si es par, False si es impar.

    return numero % 2 == 0

#Calcular el factorial de un número.
#Fórmula: n! = n * (n-1) * ... * 1

def factorial(n):

    #Parámetros:n (int)-> Número entero no negativo.
    #Retorna:int-> Factorial de n.

    if n < 0:
        raise ValueError("El factorial no está definido para números negativos.")
    resultado = 1
    for i in range(1, n+1):
        resultado *= i
    return resultado

#Calcular el promedio de una lista de números.

def promedio(lista_numeros):

    #Parámetros: lista_numeros (list)-> Lista de números (int o float).
    #Retorna:float-> Promedio de los números.

    if len(lista_numeros) == 0:
        raise ValueError("La lista no puede estar vacía.")
    return sum(lista_numeros) / len(lista_numeros)


def menu():

    print("MENÚ de opciones")
    print("1. Convertir temperatura (Celsius a Fahrenheit)")
    print("2. Calcular área de un círculo")
    print("3. Verificar si un número es par")
    print("4. Calcular factorial")
    print("5. Calcular promedio de una lista")
    print("6. Salir")
```

```python


# Bloque interactivo para solicitar valores al usuario
if __name__ == "__main__":
    while True:
        menu()
        opcion = input("Selecciona una opción: ")

        if opcion == "1":
            c = float(input("Ingresa temperatura en Celsius: "))
            print("Resultado:", celsius_a_fahrenheit(c))

        elif opcion == "2":
            r = float(input("Ingresa el radio del círculo: "))
            print("Área:", calcular_area_circulo(r))

        elif opcion == "3":
            n = int(input("Ingresa un número: "))
            print("¿Es par?:", es_par(n))

        elif opcion == "4":
            n = int(input("Ingresa un número entero: "))
            print("Factorial:", factorial(n))

        elif opcion == "5":
            lista = [float(x) for x in input("Ingresa números separados por espacio: ").split()]
            print("Promedio:", promedio(lista))

        elif opcion == "6":
            print("Saliendo del programa...")
            break

        else:
            print("Opción inválida. Intenta de nuevo.")
```
